```python
import pandas as pd
import numpy as np
completedata=pd.read_csv('StudentsPerformance.csv') 
missingdata=pd.read_csv('StudentsPerformance_miss.csv')
missingdata.info() #得到數據的特性可以看到Dtype 有些是object有些是int64
missingdata.describe()
missingdata_unique=missingdata.reading_score.unique() #使用者個函數可以叫出Column 中所有直
np.sort(missingdata_unique)
#可以發現目前的缺失值是*
```

```python
missingdata=pd.read_csv('StudentsPerformance_miss.csv',na_values="*")#使用這個將原本*的值換成nan
missingdata_unique=missingdata.reading_score.unique() 
np.sort(missingdata_unique)#可以觀察到*值已換成nan

```

```python
missingdata.isna()#可觀察到每一項的nan，並用布林值顯示
```

```python
missingdata.isna().sum()#計算布林值為true的數量有多少
#math_score/reading_score/writing_score 欄位有遺漏值
```

```python
import matplotlib.pyplot as plt
missingdata.isna().sum().plot(kind="bar")
#長條圖顯示數量多寡
```

```python
missingdata.plot(kind='scatter',title='correlationship',figsize=(8,3),x='lunch',y='reading_score',marker='o')
#相關係數圖(lunch跟reading_score的關系)
```
