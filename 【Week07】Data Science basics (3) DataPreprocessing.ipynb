### 02. Missing data處理遺漏值

##### Discussion
- As data comes in many shapes and forms, pandas aims to be flexible with regard to handling missing data. While NaN is the default missing value marker for reasons of computational speed and convenience, we need to be able to easily detect this value with data of different types: floating point, integer, boolean, and general object.

```python
# to detect missing values for an array-like object.
# Stud-performance_miss.csv
df1=pd.read_csv("Stud-performance_miss.csv")
```

```python
df1
```

```python
df1.isnull().sum()
```

```python
df1.isnull()
```

```python
df1[(df1["math score"].isnull())](df1[(df1["reading score"].isnull())])(df1[(df1["writing score"].isnull())])
```

```python
# To drop any rows that have missing data.
#df3 = df1.dropna(subset=["test01"],axis=0)
df3 = df1.dropna(axis=1)
df3
```

```python
df3 = df1.dropna(subset=["math score"])
df3
```

```python
# Filling missing data.
df2 = df1.fillna(0)
df2
```

```python
df1.fillna(df1.median())
```

```python
# Fill gaps forward or backward #bfill ffill

```

```python
# Filling with a PandasObject
```

```python
# Filling a DataFrame with the mean of that column.
df1.fillna(df1.mean())
```

### 03. Dealing with duplicate values

```python
import pandas as pd
df = pd.DataFrame({'col1':['A','A','A','A','B','B','C','C'],
                   'col2':[1,2,2,3,3,4,5,5]})
df 
```

```python
# The duplicated() function is used to indicate duplicate Series values.
# 檢查是否有重複duplicate
df.duplicated()
```

```python
df.duplicated().sum()
```

```python
# Remove Pandas series with duplicate values
# 移除重複
df.drop_duplicates()
```

```python
df.drop_duplicates(['col1','col2'])
```

```python
df.drop_duplicates(['col1'])
```

```python
df.drop_duplicates(['col1'], keep='last')
```
