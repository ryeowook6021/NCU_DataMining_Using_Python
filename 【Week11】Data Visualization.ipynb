## 1. 長條圖 Bar plot

- 使用 matplotlib.pyplot 的 bar() 方法

```python
# Resourse：http://sowf.moi.gov.tw/stat/month/list.htm

import matplotlib.pyplot as plt
import numpy as np
from pylab import mpl
plt.rcParams['axes.unicode_minus']=False
plt.style.use('ggplot')
# https://tonysyu.github.io/raw_content/matplotlib-style-gallery/gallery.html

with open('popu.txt', 'r') as fp:
    populations = fp.readlines() #__

city = list()
popu = list()

for p in populations:
    cc, pp = p.split(',')
    city.append(cc)
    popu.append(int(pp))

ind = np.arange(len(city)) #
plt.figure(figsize=(10,3))
plt.bar(ind, popu) #bar( x , y)
plt.xticks(ind,city,rotation=90)
plt.title("taiwan city population")
plt.xlabel("city")
plt.ylabel("population")

plt.show()
```

## 3. 直方圖 Histogram
- 使用 matplotlib.pyplot 的 hist() 方法

```python
import numpy as np
import matplotlib.pyplot as plt

normal_samples = np.random.normal(size = 1000) # Draw 1000 random samples from a normal (Gaussian) distribution
uniform_samples = np.random.uniform(size = 1000) # Draw 1000 samples from a uniform distribution
# A histogram is a graph showing frequency distributions.
plt.hist(normal_samples)  
plt.show()
plt.hist(uniform_samples)
plt.show()
```

### 01. Male to Female birth rate ratio in Taiwan

```python
import matplotlib.pyplot as plt
import numpy as np

with open('yrborn.txt', 'r') as fp:
    populations = fp.readlines()

yrborn = dict()

for p in populations:
    yr, tl, boy, girl = p.split()
    yrborn[yr] = {'boy': int(boy), 'girl': int(girl)}

ind = np.arange(len(yrborn))
yrlist = sorted(list(yrborn.keys()))
bp = list()
bp_b = list()
bp_g = list()

for yr in yrlist:
    boys = yrborn[yr]['boy']
    girls = yrborn[yr]['girl']
    bp.append(boys + girls)
    bp_b.append(boys)
    bp_g.append(girls)

plt.subplot(211) # 2*1: first subplot
plt.plot(bp)
plt.xlim(0,len(bp)-1)
plt.title('1986 - 2015 (Total)')

plt.subplot(212) # *1的第2張子圖
plt.plot(bp_b)
plt.plot(bp_g)
plt.xlim(0,len(bp_b)-1)
plt.title('1986 - 2015 (Boy:Girl)')

# plt.tight_layout() 
plt.show()
```

```python
import matplotlib.pyplot as pt
import numpy as np

with open('yrborn.txt', 'r') as fp:
    populations = fp.readlines()

yrborn = dict()

for p in populations:
    yr, tl, boy, girl = p.split()
    yrborn[yr] = {'boy': int(boy), 'girl': int(girl)}

ind = np.arange(1986,2016)
yrlist = sorted(list(yrborn.keys()))
bp = list()
bp_b = list()
bp_g = list()
for yr in yrlist:
    boys = yrborn[yr]['boy']
    girls = yrborn[yr]['girl']
    bp.append(boys + girls)
    bp_b.append(boys)
    bp_g.append(girls)

width = 0.35
pt.subplot(211) #2x1的第1張圖
pt.plot(ind, bp)
pt.xlim(1986,2015)
pt.title('1986 - 2015 (Total)')

pt.subplot(212)
pt.bar(ind, bp_b, width, color='b')
pt.bar(ind+0.35, bp_g, width, color='r')
pt.xlim(1986,2015)
pt.title('1986 - 2015 (Boy:Girl)')
# plt.tight_layout()
pt.show()
```
